```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

dt = pd.read_csv('q2080_2016_001_030.csv').reset_index()
dt['Year']=2016
```

```python
dt['Percentage of vehicles passing station that will charge their vehicle'] = np.round(dt['Percentage of vehicles passing station that will charge their vehicle'],3)
```

```python
OUTS = ['(O1) Mean queue length of an EV station [n]',
       '(O2) Mean waiting time in queue at an EV station [hours]',
       '(O3) Mean service time to charge at an EV station [hours]',
       '(O4) Total time spent overall at an EV station [hours]',
       '(O5) Total energy consumption of an EV station [kWh]',
       '(O6) Maximum recorded queue length of an EV station [n]',
       '(O7) Maximum waiting time in queue at an EV station [hours]',
       '(O8) Maximum time spent overall at an EV station [hours]',
       '(O9) Maximal energy consumption of an EV station [kW]']


# sns.stripplot(x='Percentage of vehicles passing station that will charge their vehicle', y=OUTS[1], data=dt[dt.Year==2026])


for COL in [0,1,2,4]:
    plt.figure(figsize=(14,7))

    g1 = sns.pointplot(x='Percentage of vehicles passing station that will charge their vehicle', y=OUTS[COL], orient='v',color='blue',data=dt[dt['Number of plugs']==1],label='1 plug')
    g2 = sns.pointplot(x='Percentage of vehicles passing station that will charge their vehicle', y=OUTS[COL], orient='v',color='red',data=dt[dt['Number of plugs']==2],label='2 plug')
    g3 = sns.pointplot(x='Percentage of vehicles passing station that will charge their vehicle', y=OUTS[COL], orient='v',color='green',data=dt[dt['Number of plugs']==7],label='7 plug')
    g4 = sns.pointplot(x='Percentage of vehicles passing station that will charge their vehicle', y=OUTS[COL], orient='v',color='brown',data=dt[dt['Number of plugs']==10],label='10 plug')
    # g.set_xticklabels(rotation=90)


    import matplotlib.patches as mpatches
    import matplotlib.pyplot as plt

    from matplotlib.patches import Patch
    from matplotlib.lines import Line2D

    P1 = Line2D([0], [0], marker='o', color='w', label='1 plug',
                              markerfacecolor='b', markersize=15)
    P2 = Line2D([0], [0], marker='o', color='w', label='2 plugs',
                              markerfacecolor='r', markersize=15)
    P3 = Line2D([0], [0], marker='o', color='w', label='7 plugs',
                              markerfacecolor='green', markersize=15)
    P4 = Line2D([0], [0], marker='o', color='w', label='10 plugs',
                              markerfacecolor='brown', markersize=15)

    plt.xticks(rotation=90)
    plt.grid()

    plt.legend(handles=[P1,P2,P3,P4])
    plt.tight_layout()
    plt.savefig('Qline{}_{}.pdf'.format(2016,str(COL)),format='pdf',bbox_inches='tight',pad_inches = 0)
```
